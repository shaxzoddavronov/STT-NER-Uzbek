Now we preprocess words and labels through aligning

```python
def align_labels_with_tokens(labels,word_ids):
  new_labels=[]
  current_word=None

  for word_id in word_ids:
    if word_id!=current_word:
      current_word=word_id
      label=-100 if word_id is None else labels[word_id]
      new_labels.append(label)

    elif word_id is None:
      new_labels.append(-100)

    else:
      label=labels[word_id]
      new_labels.append(label)

  return new_labels
```

```python
tokenized_words=tokenizer(dataset['train'][7]['words'],is_split_into_words=True)
print(tokenized_words.tokens())
```

```python
sample_labels=dataset['train'][7]['ner_tags']
sample_word_ids=tokenized_words.word_ids()
print(f"Labels: {sample_labels}")
print(f" Aligned labels: {align_labels_with_tokens(sample_labels,sample_word_ids)}")
```

```python
def tokenize_and_align_labels(samples):
    tokenized_inputs = tokenizer(
        samples["words"], truncation=True, is_split_into_words=True
    )
    all_labels = samples["ner_tags"]
    new_labels = []
    for i, labels in enumerate(all_labels):
        word_ids = tokenized_inputs.word_ids(i)
        new_labels.append(align_labels_with_tokens(labels, word_ids))

    tokenized_inputs["labels"] = new_labels
    return tokenized_inputs
```

```python
tokenized_dataset = dataset.map(tokenize_and_align_labels,
                                batched=True,
                                remove_columns=dataset["train"].column_names,)
```

```python
input_dataset=tokenized_dataset['train'].shuffle(23).train_test_split(test_size=0.1)
input_dataset['validation']=input_dataset.pop('test')
input_dataset
```

###Metrics

Most used framework to evaluate token classification is *seqeval*. With this metrics, we will compute a metric  every epoch of Trainer

```python
import evaluate
metric=evaluate.load('seqeval')
```

```python
labels = dataset["train"][0]["ner_tags"]
labels = [label_names[i] for i in labels]
preds = labels.copy()
preds[2] = "O"
preds[6] = "O"
metric.compute(predictions=[preds], references=[labels])
```

```python
import numpy as np


def compute_metrics(eval_preds):
    logits, labels = eval_preds
    predictions = np.argmax(logits, axis=-1)

    true_labels = [[label_names[c] for c in label if c != -100]
                   for label in labels]
    true_preds = [
        [label_names[p] for (p, l) in zip(prediction, label) if l != -100]
        for prediction, label in zip(predictions, labels) ]

    all_metrics = metric.compute(predictions=true_preds, references=true_labels)
    return {
        "precision": all_metrics["overall_precision"],
        "recall": all_metrics["overall_recall"],
        "f1": all_metrics["overall_f1"],
        "accuracy": all_metrics["overall_accuracy"], }
```

##Fine-Tuning Roberta with Lora Adapter

While definign model, we should pass some information of label along with model API

```python
id_to_label = {i: label for i, label in enumerate(label_names)}
label_to_id = {label:i for i, label in enumerate(label_names)}
```

We need to define Lora configuration, training arguments and trainer before fine-tuning model.

```python
import torch
from transformers import AutoModelForSequenceClassification, AutoTokenizer, TrainingArguments, Trainer
from peft import LoraConfig, get_peft_model, TaskType
from datasets import load_dataset

from transformers import AutoModelForTokenClassification
base_model=AutoModelForTokenClassification.from_pretrained(checkpoint, id2label=id_to_label, label2id=label_to_id)

# Define LoRA configuration
lora_config = LoraConfig(
    r=128,
    lora_alpha=32,
    lora_dropout=0.1,
    bias="none",
    task_type=TaskType.SEQ_CLS
)

model = get_peft_model(base_model, lora_config)
model.print_trainable_parameters()
```

The large model has over 570 million parameters, We just define ~12 million parameters as trainable

```python
training_args = TrainingArguments(
    output_dir="ShakhzoDavronov/xlm-roberta-lora-ner-uz",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    learning_rate=2e-4,
    weight_decay=0.01,
    num_train_epochs=3,
    save_total_limit=2,
    fp16=True,
    push_to_hub=True,)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=input_dataset["train"],
    eval_dataset=input_dataset["validation"],
    data_collator=data_collator,
    compute_metrics=compute_metrics,
    tokenizer=tokenizer)
```

```python
trainer.train()
```

```python
model.push_to_hub("ShakhzoDavronov/xlm-roberta-lora-ner-uz")
```

##Testing fine-tuned model (Pipeline)

```python
from peft import PeftModel, PeftConfig
from transformers import AutoModelForTokenClassification, AutoTokenizer

label_names=['O','B-PERSON','B-GPE','B-ORG','B-LOC','B-DATE','B-EVENT']
num_labels=len(label_names)
id_to_label = {i: label for i, label in enumerate(label_names)}
label_to_id = {label:i for i, label in enumerate(label_names)}

peft_model_id = "ShakhzoDavronov/xlm-roberta-lora-ner-uz"
config = PeftConfig.from_pretrained(peft_model_id)
my_model = AutoModelForTokenClassification.from_pretrained(config.base_model_name_or_path, num_labels=len(label_names),
                                                        id2label=id_to_label, label2id=label_to_id)

my_tokenizer = AutoTokenizer.from_pretrained(config.base_model_name_or_path)
my_model = PeftModel.from_pretrained(my_model, peft_model_id)
```

```python
from transformers import pipeline
ner_pipeline=pipeline('ner',model=my_model,tokenizer=my_tokenizer)
```

```python
text="""Toshkentda Shavkat Mirziyoyev Xalqaro dzyudo federatsiyasi rahbarini "Do'stlik" ordeni bilan mukofotladi."""
ner=ner_pipeline(text)
for entity in ner:
    print(entity)
```
